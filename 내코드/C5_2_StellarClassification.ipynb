```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcwd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이브 경로(C: 등) 접근 불가

# 데이터 파일 읽기 예제
import pandas as pd
X_test = pd.read_csv("./yemoonsaBigdata/datasets/Part2/stellar_X_test.csv")
X_train = pd.read_csv("./yemoonsaBigdata/datasets/Part2/stellar_X_train.csv")
y_train = pd.read_csv("./yemoonsaBigdata/datasets/Part2/stellar_y_train.csv")

# 사용자 코딩

# 답안 제출 참고
# 아래 코드 예측변수와 수험번호를 개인별로 변경하여 활용
# pd.DataFrame({'gender': pred}).to_csv('003000000.csv', index=False)
```

```python
# 1. 결측지 제거 / 사용할 컬럼 선택
#print(X_train.describe()) # 이상치 확인
#print(X_train.info()) # 결측치 확인
col_del=[]
col_num=['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift']
col_cat=['run_ID', 'rerun_ID', 'cam_col', 'field_ID', 'plate']
col_y=['galaxy']

from sklearn.preprocessing import LabelEncoder

for col in col_cat:
  le=LabelEncoder()
  le.fit(X_train[col])
  X_train[col]=le.transform(X_train[col])
  le.fit(X_test[col])
  X_test[col]=le.transform(X_test[col])
```

```python
# 2. 데이터 스플릿 / 스케일링
from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = train_test_split(X_train[col_num],y_train,test_size=0.3)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_tr[col_num])
x_tr[col_num]=scaler.transform(x_tr[col_num])
x_val[col_num]=scaler.transform(x_val[col_num])
X_test[col_num]=scaler.transform(X_test[col_num])
```

```python
# 3. 모델링 / 평가 (ROC_AUC)
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(x_tr,y_tr)
pred=rf.predict_proba(x_val)

from sklearn.metrics import roc_auc_score

score=roc_auc_score(y_val,pred[:,1])
score
```

```python
from xgboost import XGBClassifier
rf=XGBClassifier()
rf.fit(x_tr,y_tr)
pred=rf.predict(x_val)

from sklearn.metrics import roc_auc_score,accuracy_score

score=roc_auc_score(y_val,pred)

score
```

```python
from lightgbm import LGBMClassifier
rf=LGBMClassifier()
rf.fit(x_tr,y_tr)
pred=rf.predict(x_val)

from sklearn.metrics import roc_auc_score

score=roc_auc_score(y_val,pred)
score
```
