#1.Iris 데이터 세트

```python
import pandas as pd
data = pd.read_csv("./yemoonsaBigdata/datasets/Part2/iris.csv")
data

pd.set_option('max_rows',None)
pd.set_option('max_columns',None)
data
```

##1) Iris 데이터 세트에서 species 변수 값별로 petal_width 상위 25% 값을 구한 후, 가장 큰 값과 가장 작은 값 사이의 차이를 계산하시오

```python
d=data.groupby('species')['petal_width'].describe()
print(d)
print(d.iloc[0:3,6].max()-d.iloc[0:3,6].min())
```

```python
d=data.groupby('species')['petal_width'].quantile(0.75)
print(round(d.max()-d.min()))
```

##2) Iris 데이터 세트에서 sepal_length 변수의 전체 값들을 중복 없이 리스트로 만든 후, 오름차순으로 정렬했을 때 리스트의 10번째에 오는 숫자를 계산하시오.

```python
print(data['sepal_length'].drop_duplicates().sort_values(ascending=True).tolist()[9])
print(sorted(data['sepal_length'].drop_duplicates().tolist())[9])
print(sorted(data['sepal_length'].unique().tolist())[9])
```

```python
ds=sorted(data['sepal_length'].unique().tolist())
print(ds[9])
print(type(ds))
```

```python
a=data['sepal_length'].unique().tolist()
a.sort()
print(a[9])
```

##3) Iris 데이터 세트를 sepal_width 변수 값으로 내림차순한 후 위에서부터 100개 행을 학습 데이터로 분리하고, 학습 데이터에서 sepal_width의 표준편차 값을 구한 후 50을 곱한 값을 계산하시오(최종 계산 값의 소수점 이하는 버린다.)

```python
import math 
td=data.sort_values(by='sepal_width',ascending=False).head(100)
print(math.trunc(td['sepal_width'].std()*50))
```

##4) Iris 데이터 세트에서 species 변수 값별로 petal_length 상위 10개 행(큰 값)의 평균값을 구한 후, 평균값의 합을 계산하시오(최종 계산 값의 소수점 이하는 반올림한다.)

```python
a=data.sort_values(by='petal_length',ascending=False).groupby('species')['petal_length'].head(10)[0:9].mean()
b=data.sort_values(by='petal_length',ascending=False).groupby('species')['petal_length'].head(10)[10:19].mean()
c=data.sort_values(by='petal_length',ascending=False).groupby('species')['petal_length'].head(10)[20:29].mean()
print(round(a+b+c))
```

```python
round(sum(data.sort_values(by='petal_length',ascending=False).groupby('species').head(10).groupby('species')['petal_length'].mean()))
```

##5) Iris 데이터 세트에서 sepal_length 변수가 이상치를 가지는 데이터 행 수를 계산하시오(단, 이상치 범위는 (평균＋2×표준편차) 이상이거나 (평균－2×표준편차) 이하이다.)

```python
aas = data['sepal_length'].mean() + 2*data['sepal_length'].std()
bs = data['sepal_length'].mean()-2*data['sepal_length'].std()

print(data[(data['sepal_length']>=aas) | (data['sepal_length']<=bs)].count())
print(len(data[(data['sepal_length']>=aas)|(data['sepal_length']<=bs)]))
```

#2. Wine 데이터 세트

```python
import pandas as pd
data = pd.read_csv("./yemoonsaBigdata/datasets/Part2/wine.csv")
data.info()
```

##1) Wine 데이터 세트에서 total_phenols 결측치를 제거하고 alcalinity_of_ash 변수의 결측치는 중앙값으로 대체한 후, alcalinity_of_ash 평균값을 계산하시오(최종 계산 값의 소수점 이하는 버린다.).

```python
import math
dt=data
dt['total_phenols']=data['total_phenols'].dropna()
print(math.trunc(data['alcalinity_of_ash'].fillna(data['alcalinity_of_ash'].median()).mean()))
```

##2) Wine 데이터 세트에서 alcohol 값이 alcohol 평균 값 이상이고, color_intensity 값이 color_intensity 평균 값 이상인 데이터만 남겼을 때, 가장 많은 class 값을 구하시오

```python
 a1=data['alcohol']>=data['alcohol'].mean()
 a2=data['color_intensity']>=data['color_intensity'].mean()
data[a1&a2]['class'].value_counts().max()
```

###3) Wine 데이터 세트에서 class 1에 해당하는 데이터의 ash 평균값과 Wine 데이터 세트의 모든 결측치를 제거한 후의 ash 값의 평균값의 차이의 절댓값을 계산하시오(최종 계산 값의 소수점 아래 4번째 자리에서 반올림한다.)

```python
abs(round(data[data['class']==1]['ash'].mean()-data.dropna()['ash'].mean(),3))
```

##4) Wine 데이터 세트에서 alcohol 변수의 상위에서 20번째 값(alcohol을 오름차순 정렬했을 때 20번째에 위치한 값) 으로 상위 20개의 값을 변환한 후, proanthocyanins가 2 이하인 데이터를 추출하여 alcohol의 평균값을 계산하시오(최종 계산 값의 소수점 아래 3번째 자리에서 반올림한다.).

```python
a=data.sort_values(by='alcohol',ascending=True).reset_index()
b=a.loc[19,'alcohol']
a.loc[:19,'alcohol']=b

round(a[a['proanthocyanins']<=2]['alcohol'].mean(),2)
```

##5) Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 제 3사분위수와 제 1사분위수 차를 구하고, class 2에 해당하는 데이터의 proline 변수의 제 3사분위수와 제 1사분위수 차를 구하여 두 값의 차이의 절댓값을 계산하시오

```python
a=data[data['class']==1]['proline'].quantile(0.75)-data[data['class']==1]['proline'].quantile(0.25)
b=data[data['class']==2]['proline'].quantile(0.75)-data[data['class']==2]['proline'].quantile(0.25)
abs(a-b)
```
